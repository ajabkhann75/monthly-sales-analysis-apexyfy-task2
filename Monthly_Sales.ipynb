
### Objective
The objective of this task is to visualize monthly sales data using a line chart to understand the sales trend over a year.

### Tools Used
- Python
- Pandas
- Matplotlib

### Dataset
- Month
- Sales
---

#### Step 1: Import Libraries 

```python
import pandas as  pd
import matplotlib.pyplot as plt
```

#### Step 2: Load Dataset 

```python
df=pd.read_csv("Apexyfy_Monthly_sales_Analysis.csv")
```

#### Step 2: Display the Data

To verify that the dataset is loaded correctly, the first few rows are displayed.

```python
df.head()
```

#### step 3: 
- Used the `plot()` function to create a line chart of sales against months.
- Labeled the X-axis as "Month" and Y-axis as "Sales" for better understanding.
- added a title "Monthly Sales Trend" to the graph.
- Displayed the final chart using `plt.show()` to visualize the sales trend.

```python
plt.plot(df["Month"],df["Sales"])

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

plt.show()
```

## Conclusion

- In this task, monthly sales data was successfully visualized using Python, Pandas, and Matplotlib. The dataset was analyzed by plotting a line chart to represent sales trends over a 12-month period.

- The visualization clearly highlighted patterns in sales performance, showing both growth and fluctuations throughout the year. Peak sales were observed during certain months, while a noticeable decline occurred in others, indicating possible seasonal or market-driven variations.

- Overall, this task demonstrated how data visualization techniques can be effectively used to identify trends, analyze business performance, and support better decision-making based on data insights.
